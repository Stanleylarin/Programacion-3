```python
import zipfile
import io
import tensorflow as tf
import math
import tensorflow_datasets as tfds
import cv2
import numpy as np
import os 
import matplotlib.pyplot as plt
```

```python
ruta_ropa='/content/Ropa'
datos=[]
img_size=150

for img in os.listdir(ruta_ropa):
  img = cv2.imread(os.path.join(ruta_ropa,img))
  img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  img_gray_resize = cv2.resize(img_gray,(img_size,img_size))
  datos.append([img_gray_resize])
```

```python
plt.figure()
plt.imshow(np.squeeze(datos[4]))
plt.colorbar()
plt.grid(True)
plt.show()
```

```python
images = np.concatenate([datos])
print(images.shape)
```
